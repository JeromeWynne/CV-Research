```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Variation in self-information with outcome probability

```python
x = np.arange(0.01, 1, 0.01)
I = -np.log(x)

plt.figure(figsize = (15, 4))
plt.plot(x, I);
plt.xlabel('Probability'); plt.ylabel('Information content');
```

### Self-information of normally distributed outcomes

```python
def dnorm(x, mu, sigma):
    return np.exp(- 0.5 * (x - mu)**2 / sigma**2) / (2*np.pi*sigma**2)

x = np.arange(-5, 5, 0.05)
MU = 0; SIGMA = 1;
f, axs = plt.subplots(1, 2, figsize = (15, 5))
axs[0].plot(x, dnorm(x, MU, SIGMA)); axs[0].set_title('Normal probability distribution');
axs[1].plot(x, -np.log(dnorm(x, MU, SIGMA))); axs[1].set_title('Parabolic self-information');
for mu, sigma in zip([0]*10, np.arange(1, 2, 0.1)):
    axs[0].plot(x, dnorm(x, mu, sigma), color = plt.cm.viridis(sigma-1))
    axs[1].plot(x, -np.log(dnorm(x, mu, sigma)), color = plt.cm.viridis(sigma-1))
```

### Expected self-information for the outcomes of a binary variable
*Expected self-information = Shannon entropy*

```python
p = np.arange(0.01, 1, 0.01)
plt.plot(p, -(p*np.log(p) + (1-p)*np.log(1-p)))
plt.xlabel('Success probability'); plt.ylabel('Average outcome self-information (Shannon entropy)');
```

### Kullback-Leibler Divergence of two normal distributions

```python
def KL(p, q):
    return p*np.log(p) - p*np.log(q)
```

```python
x         = np.arange(-5, 5, 0.05)
true_mean = 0
true_sd   = 1
p         = dnorm(x, true_mean, true_sd)
estimated_mean = -1
estimated_sd   = 1.5
q              = dnorm(x, estimated_mean, estimated_sd)

f, axs = plt.subplots(1, 2, figsize = (15, 5))
axs[1].plot(x, KL(q, p), label = 'Terms of KL(q, p)');
axs[1].plot(x, KL(p, q), label = 'Terms of KL(p, q)');
axs[1].legend(loc = 3)
axs[0].plot(x, p, label = 'p'); 
axs[0].plot(x, q, label = 'q'); 
axs[0].legend(loc = 3);
```
