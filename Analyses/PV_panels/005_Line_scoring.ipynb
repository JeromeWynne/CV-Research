This notebook demos a pair of functions that I considered for detecting lines in the grids of solar panels.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
W = 100
x = np.arange(1, W)

a     = np.arange(1, 20, 1)
GAMMA = np.log(a)/np.log(np.max(a))
GAMMA = GAMMA[1:-1]
c_idx = np.linspace(1, 0.2, GAMMA.shape[0])

plt.figure(figsize = (12, 6))

for i, gamma in zip(c_idx, GAMMA):
    plt.plot(x, (1 - gamma**(W-x)), color = plt.cm.hot(1-i), linewidth=1)
    plt.xlabel('Number of zero pixels (of W = 100)', fontsize = 14)
    plt.ylabel('Line score', fontsize = 14)
```

```python
W = 100
x = np.arange(1, W)

a     = np.arange(0, 4, 0.25)
BETA  = np.exp(a)
c_idx = np.linspace(0.9, 0, BETA.shape[0])

plt.figure(figsize = (12, 6))

for i, beta in zip(c_idx, BETA):
    plt.plot(x, (x/W)**beta, color = plt.cm.viridis(i), linewidth=1)
    plt.xlabel('Number of non-zero pixels (W+ out of W = 100)', fontsize = 14)
    plt.ylabel('Line score', fontsize = 14)
    
plt.savefig('line_detector_score.pdf')
```
